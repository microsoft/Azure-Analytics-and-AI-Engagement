```python
def readConfigFile(filePath):
    rdd = spark.sparkContext.wholeTextFiles(TransformationConfigFolder + '/' + filePath)
    return rdd.collect()[0][1]
```

##  Transform data to given target schema.

```python
dmf_adaptor_content = readConfigFile('Adapter.json')
target_db_semantics_content = readConfigFile('DBSemantics.json')
target_db_semantics_config_content = readConfigFile('DBSemanticsConfig.json')
target_db_schema_content = readConfigFile('DBTargetSchema.json')
db_schema_config_content = readConfigFile('DBTargetSchemaConfig.json')
env_config_content = readConfigFile('EnvConfig.json')


spark.conf.set("spark.sql.legacy.timeParserPolicy", "LEGACY")
spark.conf.set("spark.sql.sources.partitionOverwriteMode", "DYNAMIC")
spark.conf.set("spark.sql.caseSensitive", "true")

{
    DMFApplication
        .start_with_configuration_content(spark= spark,
                                            dmf_adaptor= dmf_adaptor_content,
                                            target_db_semantics= target_db_semantics_content,
                                            target_db_semantics_config= target_db_semantics_config_content,
                                            target_db_schema= target_db_schema_content,
                                            db_schema_config= db_schema_config_content,
                                            env_config= env_config_content)
}
```
