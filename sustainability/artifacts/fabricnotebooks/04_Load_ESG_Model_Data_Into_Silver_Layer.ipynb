```python

table_name = 'Dim_SGCSRD_Reporting_Year'

def LoadFullDataFromSource(Table_Name):
  file_location = "abfss://#WORKSPACE_ID#@onelake.dfs.fabric.microsoft.com/#BRONZE_LAKEHOUSE_ID#/Files/TablesCSV/Reporting_Year.csv"
  df = spark.read.format("csv").option("header","true").load(file_location) 
  df.show(10) 
  df = df.withColumn("ReportingYear", to_date(unix_timestamp(col('ReportingYear'), 'yyyy')))
  df = df.withColumn('ReportStartDate', to_date(unix_timestamp('ReportStartDate','yyyy-MM-dd').cast("date")))
  df = df.withColumn('ReportEndDate', to_date(unix_timestamp('ReportStartDate','yyyy-MM-dd').cast('date')))
  df.show(10) 
  df.write.mode("overwrite").format("delta").save("Tables/" + Table_Name)
      
LoadFullDataFromSource(table_name)
```
