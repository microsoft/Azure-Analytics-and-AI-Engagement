# Importing the necessary libraries

```python
import os
from pyspark.sql.functions import *
from pyspark.sql.functions import col, unix_timestamp, to_date
from pyspark.sql.types import DateType
spark.conf.set("sprk.sql.parquet.vorder.enabled", "true")
spark.conf.set("spark.microsoft.delta.optimizeWrite.enabled", "true")
spark.conf.set("spark.microsoft.delta.optimizeWrite.binSize", "1073741824")
```

# List of tables

```python
DimTablesList = ['dimension_brand','dimension_city','dimension_customer','dimension_date','dimension_country','dimension_product','dimension_employee']
FactTablesList = ['fact_campaigndata','fact_order','fact_store_sales_data']
```

## Shortcut Folder Path

```python
DimPath='abfss://#WORKSPACE_NAME#@onelake.dfs.fabric.microsoft.com/lakehouseBronze.Lakehouse/Files/data/DimensionData/'
FactPath='abfss://#WORKSPACE_NAME#@onelake.dfs.fabric.microsoft.com/lakehouseBronze.Lakehouse/Files/data/FactData/'
```

## Reading the data from source csv

```python
def LoadFullDimDataFromSource(Table_Name):
  file_location =Table_Name
  if "dimension_product" in Table_Name:
    df = spark.read.format("csv").option("header","true").load(DimPath+Table_Name)  
    df=df.withColumn('TargetGeneration', regexp_replace('TargetGeneration', 'Boomer; GenX; GenY; GenZ', 'Boomers'))
    df.write.mode("overwrite").format("delta").option("header", "true").saveAsTable("lakehouseSilver." + Table_Name)
    print(Table_Name + " data loaded successfully in Lakehouse..!!!")
  else:
    df = spark.read.format("csv").option("header","true").load(DimPath+Table_Name) 
    df.write.mode("overwrite").format("delta").option("header", "true").saveAsTable("lakehouseSilver." + Table_Name)
    print(Table_Name + " data loaded successfully in Lakehouse..!!!")
```

```python
def LoadFullFactDataFromSource(Table_Name):
  file_location =Table_Name
  if "fact_campaigndata" in Table_Name:
    df = spark.read.format("csv").option("header","true").load(FactPath+Table_Name)  
    df = df.withColumn("ProductCategory",col("ProductCategory").cast("string"))
    df = df.withColumn("CampaignID",col("Campaign_ID").cast("integer"))
    df = df.withColumn("Campaign_ID",col("Campaign_ID").cast("integer"))
    df = df.withColumn("ROI",col("ROI").cast("integer"))
    df = df.withColumn("Revenue_Target",col("Revenue_Target").cast("Double"))
    df = df.withColumn("Cost",col("Cost").cast("Double"))
    df = df.withColumn("Revenue",col("Revenue").cast("Double"))
    df = df.withColumn("Responses",col("Responses").cast("Double"))
    df = df.withColumn("Profit",col("Profit").cast("Double"))
    df = df.withColumn("Marketing_Cost",col("Marketing_Cost").cast("Double"))
    df.write.mode("overwrite").format("delta").option("header", "true").saveAsTable("lakehouseSilver." + Table_Name)
    print(Table_Name + " data loaded successfully in Lakehouse..!!!")

  else:
    df = spark.read.format("csv").option("header","true").load(FactPath+Table_Name) 
    df.write.mode("overwrite").format("delta").option("header", "true").saveAsTable("lakehouseSilver." + Table_Name)
    print(Table_Name + " data loaded successfully in Lakehouse..!!!")
```

# Calling the user defined function

```python
for table in DimTablesList:
    LoadFullDimDataFromSource(table)
for table in FactTablesList:
    LoadFullFactDataFromSource(table)
```
