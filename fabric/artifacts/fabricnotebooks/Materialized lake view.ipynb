```python
%%sql



   CREATE MATERIALIZED LAKE VIEW IF NOT EXISTS dbo.cleaned_order_data AS
   SELECT 
       o.TransactionId,
       o.SaleDate,
       o.Product_ID,
       p.ProductName,
       o.Quantity,
       p.BasePrice,
       o.Quantity * p.BasePrice AS revenue
   FROM Snowflake.df_orders o
   JOIN AWS.dim_product p
   ON o.Product_ID = p.Product_ID;
```

```python
%%sql


   CREATE MATERIALIZED LAKE VIEW IF NOT EXISTS dbo.product_sales_summary AS
   SELECT
       Product_ID,
       ProductName,
       SUM(Quantity) AS total_quantity_sold,
       SUM(revenue) AS total_revenue,
       ROUND(AVG(revenue), 2) AS average_order_value
   FROM
       dbo.cleaned_order_data
   GROUP BY
       Product_ID,
       ProductName;
```
