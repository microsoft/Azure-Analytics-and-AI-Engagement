```sql
-- CommunicationDevice

CREATE TABLE dbo.CommunicationDevice (
    CommunicationDeviceId bigint,
    CommunicationDeviceImei bigint,
    CommunicationDeviceName varchar(5000),
    CommunicationDeviceDescription varchar(5000),
    CommunicationDeviceManufacturerName varchar(5000),
    CommunicationDeviceTypeId int,
    CommunicationDeviceModel varchar(5000),
    FirmwareVersion varchar(5000),
    SerialNumber varchar(5000),
    MacAddress varchar(5000),
    MobileStationClassMarkId int,
    BatteryTechnologyTypeId int,
    WirelessInterfaceId int,
    MobileStationClassId int,
    MobileStationTypeId int,
    ProcessorClockSpeedGhz decimal(18, 2),
    MaximumMemoryGb decimal(18, 2),
    STmsi varchar(5000),
    Mmec varchar(5000),
    MTmsi varchar(5000),
    LengthCm decimal(18, 2),
    WidthCm decimal(18, 2),
    HeightCm decimal(18, 2),
    WeightKg decimal(18, 2),
    MinimumExternalOperatingTemperatureC decimal(18, 2),
    MaximumExternalOperatingTemperatureC decimal(18, 2),
    EstimatedMtbfHours decimal(18, 2),
    MinimumOperatingHumidityPercent decimal(18, 2),
    MaximumOperatingHumidityPercent decimal(18, 2),
    BackupBatteryTime decimal(18, 2),
    PowerConsumptionTypicalWHr decimal(18, 2),
    Color varchar(5000),
    DisplayDiagonalSizeCm decimal(18, 2),
    DisplayResolutionPixels int,
    BatteryCapacityMah decimal(18, 2),
    Note varchar(5000)
);
```
